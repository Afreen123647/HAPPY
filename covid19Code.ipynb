```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
import math
data=pandas.read_csv('covid19.csv')
data.head()
```

```python
data.drop('Sno',axis=1,inplace=True)
data
```

```python
dataG = data.groupby('Date',axis=0,sort=False)
df=dataG.sum()
df
```

```python
df["Total"]=df.sum(axis=1)
df.head()
```

```python
plt.plot(df)
plt.title("national covid 19 cases")
plt.xlabel("Dates")
plt.ylabel("No. of cases")
plt.show()
```
